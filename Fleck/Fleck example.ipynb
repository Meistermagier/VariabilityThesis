```python
import matplotlib.pyplot as plt
import numpy as np
import fleck
from fleck import Star
from fleck import generate_spots
import astropy.units as u
```

```python
from fleck import generate_spots

spot_contrast = 0.7
u_ld = [0.5079, 0.2239]

n_phases = 30
n_inclinations = 100
n_spots = 3

spot_radius = 0.1   # Rspot/Rstar
min_latitude = 70   # deg
max_latitude = 90   # deg

lons, lats, radii, inc_stellar = generate_spots(min_latitude, max_latitude,
                                                spot_radius, n_spots,
                                                n_inclinations=n_inclinations)
```

```python
star = Star(spot_contrast=spot_contrast, n_phases=n_phases, u_ld=u_ld)
lcs = star.light_curve(lons, lats, radii, inc_stellar)
```

```python
plt.plot(star.phases, lcs)
plt.show()
```

```python
inc_stellar = 90 * u.deg
spot_radii = np.array([[0.1], [0.24],[0.18]])
spot_lats = np.array([[-10], [15], [25]]) * u.deg
spot_lons = np.array([[360-30], [80], [0]]) * u.deg

times = np.linspace(0, 4, 500)
u_ld = [0.5079, 0.2239]

star = Star(spot_contrast=0.7, u_ld=u_ld, rotation_period=0.5,n_phases=500)

ax = star.plot(spot_lons, spot_lats, spot_radii, inc_stellar,time=0)
```

```python

lc = star.light_curve(spot_lons,spot_lats,spot_radii,inc_stellar,times=times)
plt.plot(times,lc)
```
